```python
import numpy as np

# 生成6行4列的元素为 0-23 的 ndarray，记为 t1
t1 = np.arange(24).reshape((6,4))
print('ndarray类型的数组：\n',t1)

# 将该 ndarray 转成列表赋给 t2
t2=t1.tolist()
print('list 型的数组：\n',t2)

# ndarray 可以直接和整数进行加减乘除运算，但是 list 不行
# 注意：这些运算并不会改变原来的 ndarray 数组本身
print('ndarray + 2 的结果：\n',t1+2)
print('ndarray - 2 的结果：\n',t1-2) 
print('ndarray * 2 的结果：\n',t1*2)
print('ndarray / 2 的结果：\n',t1/2)

# print(t2+2)  # 不能对列表进行直接加整数操作
```

## ndarray数组之间的运算
    形状相同才能运算
    运算结果为对应位置元素相加/乘
    注不是矩阵乘法
    这里是对应位置元素相乘意：两个 ndarray 相乘，矩阵是一行乘一列，并且矩阵乘法6行4列只能和4行6列的相乘
    要想得到矩阵乘法的结果要调用 dot()接口（点乘）
    
    shape不同，不能进行运算
    
    但一维数组和二维数组可以进行运算，前提是一维的元素个数和二位数组的列数相等
    一维数组只要列和二维数组相等，就会将一维数组作用到二位数组的每一行上

```python
import numpy as np

# 形状相同,才可以进行加减乘除
t1 = np.arange(24).reshape((6,4))
t2 = np.arange(100,124).reshape((6,4))
print(t1)
print(t2)
print(t1+t2) 
print(t1*t2)  #是否是矩阵乘法？ 不是，这里是对应位置元素相乘，矩阵是一行乘一列，并且矩阵乘法6行4列只能和4行6列的相乘



```

```python
# shape不同，不能进行运算
# t1 = np.arange(24).reshape((4,6)) 
# t2 = np.arange(6).reshape((2,3)) 
# print(t1)
# print(t2) 
# print(t1-t2) 
```

```python
# 但一维数组和二维数组可以进行运算，前提是一维的元素个数和二位数组的列数相等
t1 = np.arange(24).reshape((4,6)) 
t2 = np.arange(0,6)
print(t1)
print(t2) 
print(np.sum(t2))  # np.sum()会对ndarray中的元素求和
print(t2.shape)
print(t1-t2)  # 一维数组只要列和二维数组相等，就会将一维数组作用到二位数组的每一行上
```

```python
# 一行多列的可以，多行一列的其实也可以
# 多维数组只要只有一列，就会将这一列作用到二位数组的每一列上
t1 = np.arange(24).reshape((4,6)) 
t2 = np.arange(4).reshape((4,1))
print(t1)
print(t2)
print(t1-t2)
```

### 数组按指定轴相加
    按某个轴相加，就是在那个轴方向上的所有对应位置元素相加
    

```python
import numpy as np

a = np.array([[1,2,3],[4,5,6]])
print(a)

# 按 0轴 求和（0轴是最外层，按照从第一行到最后一行的方向，也就是每一列求和）
print(np.sum(a, axis=0))  # [5 7 9]
print("-"*20)
# 按 1轴 求和
print(np.sum(a,axis=1))  # [ 6 15]
print("-"*20)
# ndayyay所有元素求和，得到的是一个数
print(np.sum(a))

a=np.arange(27).reshape((3,3,3))
print(a)
print("-"*20)

b=np.sum(a, axis=0) # 三维的，按轴0相加，最外层的方向，就是每个二维数组对应位置元素相加
print("-"*20)
print(b)

c=np.sum(a, axis=1) # 三维的，按轴1相加，倒数第二层方向，就是每个二维数组内部的每一列各元素相加
print("-"*20)
print(c)

d=np.sum(a, axis=2) # 三维的，按轴2相加，倒数第三层方向，就是每个二维数组内部的每一行各元素相加
print("-"*20)
print(d)
```

# 五、数组的索引和切片

## 一维数组索引和切片
    冒号分隔切片参数 start:stop:step 
    例：print(a[2:7:2]) 从索引 2 开始到索引 7 停止，间隔为 2

```python
import numpy as np

a = np.arange(10)

# 如果只放置一个参数，如 [2]，将返回将该索引作为下表的数组元素
print(a,'\n',a[2])

# 如果为 [2:]，表示从该索引开始以后的所有项都将被提取
print(a[2:])
```

## 二维数组索引和切片
    numpy 对二位数组的切片操作要比 python 高级
    它不但能切取某一行，还能取某一列，不仅可以对行切片，还能对列切片
    其实思想非常简单
    array0[start1:stop1:step1 , start2:stop2:step2]  前面对行切片，后面对列切片，中间逗号隔开
    想要取不连续的行或者列，就把start:stop:step 写成一个列表, 比如[0,2,3] 就是取第0,2,3列

```python
import numpy as np

t1 = np.arange(24).reshape(4,6)
print(t1)
print('*'*20)

print(t1[1]) # 取一行(一行代表是一条数据，索引也是从0开始的) 
print('*'*20)

print(t1[1:]) # 取连续的多行
print('*'*20)

print(t1[1:3:]) # 取连续的多行
print('*'*20)

print(t1[[0,2,3]]) # 取不连续的多行，两个方括号，取到的是第0行、第2行、第三行
print('*'*20)

print(t1[[0,2,3],:]) # 取不连续的多行，就比上面的方法多写个逗号和冒号，逗号冒号可写可不写
print('*'*20)

print(t1[:,-1]) # 取一列,-1就是倒着取
print('*'*20)

print(t1[:,1:]) # 连续的多列
print('*'*20)

print(t1[:,[0,2,3]]) # 取不连续的多列
print('*'*20)

print(t1[2,3]) # 取某一个值,三行四列  py是t1[2][3]
print('*'*20)

print(t1[[0,1,1],[0,1,3]])# 取多个不连续的值，[[行，行。。。],[列，列。。。]]
```

# 六、数组中的数值修改
    注意：这是在数组原空间上的操作


```python
import numpy as np

t = np.arange(24).reshape(4,6)
print(t)
print('-'*50)

# 修改某一行的值（批量赋值）
# t[1,:]=0  

# 修改某一列的值（批量赋值）
# t[:,1]=0

# 修改连续多行（切片+批量赋值）
# t[1:3,:]=0

# 修改连续多列（切片+批量赋值）
# t[:,1:4]=0

# 修改多行多列，交叉处批量赋值
# t[1:3,2:5]=0

# 修改多个不相邻的点
# t[[0,1],[1,3]]=0  # 点(0,1)和(1,3)赋值为0

# 可以根据条件修改，比如讲小于10的值改掉
# t[t<10]=0

# 使用逻辑判断：np.logical_and与(&)    np.logical_or或(|)   np.logical_not非(~)
# t[(t>2)&(t<6)]=0  # 与
# t[(t<2)|(t>6)]=0  # 或
# t[~(t>6)]=0  # 非

# 财富平均化：将所有人的财富都拉到10与18之间，小于10的改成10，大于18的改成18
t=t.clip(10,18)

print(t)
print('-'*50)

# 拓展：三目运算 -- np.where(condition, x, y)满足条件condition，输出x，不满足输出y。
score = np.array([[80,88],[82,81],[75,81]])
print(score)
result = np.where(score<80,True,False) # <80分的赋True,>=80分的赋False
print(result)
score[result]=100  # 只会改变True位置的元素
print(score)
```

# 七、数据的添加、删除、去重
### 末尾添加 - append()

    numpy.append() 会在数组的末尾添加值
    append 操作会new一个新的数组，并把原来的数组所有元素赋值过去
    原本的数组并不会因为 append() 操作而发生改变
    此外，输入的数组维度必须匹配，否则会报错 ValueError
    
    参数说明：
    arr:  输入数组
    values: 要向arr添加的值，需要和 arr形状相同（除了要添加的轴）
    axis: 默认为None。
        当axis无定义时，是横向加成，返回总是为一维数组!
        当axis有定义的时候，分别为0和1的时候（列数要相同）。当axis为1时，数组是加在右边（行数要相同）。

```python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])  # 新建二维 ndarray

print('第一个数组：')
print(a)
print('\n')

print('不指定轴向数组添加元素:')
print(np.append(a, [7, 8, 9]))  # 这里直接append(a, [7, 8, 9])没有指定维度，结果将会是一个一维数组
print('\n')
print(a)  # 这里的 a 没变
print('-'*50)

b=np.array([[7, 8, 9]])
print(b.ndim)
print(b)
print('-'*50)

print('沿轴 0 添加元素：')
print(np.append(a, [[7, 8, 9]], axis=0))  # 注意有两层中括号，并且要指定轴0
print('\n')
print(a)  # a 还是没有变

print('沿轴 1 添加元素：')
print(np.append(a, [[5, 5, 5], [7, 8, 9]], axis=1))
print(a)  # a 依旧没有变
print('-'*50)
```

### 指定位置添加&广播 - insert() 
    np.insert(a,1, [11, 12],axis=0)  
    在给定索引(1)之前，沿给定轴(axis=0)在数组(a)中插入值。
    
    如果值的类型转换为要插入，则它与输入数组不同。
    插入没有原地的，函数会返回一个新数组。 
    
    如果未提供轴，则输入数组会先被展开，展开后再插入。
    
    广播的概念：
    在给定轴的方向上，在每个平行元素上相同位置填充值，直到符合形状（我新增了x元素，我的所有兄弟都要新增x）

```python
import numpy as np

a = np.array([[1, 2], [3, 4], [5, 6]])

print('第一个数组：')
print(a)
print('\n')

print('未传递 axis 参数,在插入 之前 输入数组会被展开：')  # 注意：是先展开了，再按展开后的形状（一维）插入的
print(np.insert(a,1, [11, 12]))
print('\n')

print('传递了 axis 参数,会广播值数组来匹配输入数组：')  
print(np.insert(a,1, [11, 12],axis=0))


print('沿轴 0 广播：')
print(np.insert(a, 1, 11, axis=0))  # 11会在轴0方向（）
print('\n')

print('沿轴 1 广播：')
print(np.insert(a, 1, [11], axis=1))
```

## 删除 - delete()
    numpy.delete() 函数返回从输入数组中删除指定子数组的新数组。 与 insert() 函数的情况一样，
    如果未提供轴参数， 则输入数组将展开。
    不改变原数组
    
    参数说明 ： np.delete(a,1,axis = 0)
    a： 输入数组
    obj：可以被切片，整数或者整数数组，表明要从输入数组删除的子数组
    axis：沿着它删除给定子数组的轴，如果未提供，则输入数组会被展开

```python
import numpy as np
a = np.arange(12).reshape(3,4)

print('第一个数组：')
print(a)
print()

print('未传递 axis 参数。 在删除 之前 输入数组会被展开：')
print(np.delete(a,5))
print()

print('原数组不会被改变：')
print(a)  # 不改变原数组
print()

print('删除下标为1的行：')
print(np.delete(a,1,axis = 0))
print('\n')

print('删除下标为1的列：')
print(np.delete(a,1,axis = 1))
```

### 去重 - unique() 
    numpy.unique() 函数用于去除数组中的重复元素。
    不会改变原数组，去重后自动排序
    
    参数说明：u,indices = np.unique(a, return_index = True)
    arr：输入数组，如果不是一维数组则会展开
    return_index：如果为true，返回新列表元素在旧列表中的位置（下标），并以列表形式存储
              其实就是去重后的元素在原数组中第一次出现的下标值组成的数组
    return_inverse：如果为true，返回旧列表元素在新列表中的位置（下标），并以列表形式存储
    return_counts：如果为true，返回去重数组中的元素在原数组中的出现次数，并以列表形式存储

```python
import numpy as np
a = np.array([5,2,6,2,7,5,6,9,8,2])

print ('第一个数组：')
print (a)
print ('\n')

#去除重复元素，去重后自动排序
print ('第一个数组的去重值：')
u = np.unique(a)
print (u)
print('不会改变原数组：')
print(a)
print ('\n')

# return_index 参数
u,indices = np.unique(a, return_index = True)
print('去重后的数组：')
print(u)
print ('去重后的数组在原数组中的索引组成的数组：')
print (indices)
print ('\n')

# return_inverse 参数
u,indices = np.unique(a,return_inverse = True)
print('去重后的数组：')
print (u)
print('原数组在去重后的数组中的索引组成的数组：')
print (indices)
print ('\n')

# return_counts 参数
u,indices = np.unique(a,return_counts = True)
print('去重后的数组：')
print (u)
print ('去重元素的重复次数组成的数组：')
print (indices)
```

# 八、numpy 的计算

        
    拓展：方差var, 协方差cov, 计算平均值 average, 计算中位数 median
    方差：每个数减去平均数平方再求和，最后除以数的个数
    标准差：方差开根号

```python
import numpy as np

score = np.array([[80,88],[82,81],[75,81]])
print('原数组：','\n',score)

# 1. 获取所有数据最大值
result = np.max(score)
print('数组中的最大值：',result)

# 2. 获取某一个轴方向上的每一个平行数组中数据最大值
result = np.max(score,axis=1)
print(result)

# 3. 获取最小值
result = np.min(score)
print(result)

# 4. 获取某一个轴方向上的每一个平行数组中数据最小值
result = np.min(score,axis=0)
print(result)

# 5. 数据的比较
result = np.maximum( [-2, -1, 0, 1, 2], 0 )  # 第一个参数中的每一个数与第二个参数比较返回大的
print(result)
result = np.minimum( [-2, -1, 0, 1, 2], 0 )  # 第一个参数中的每一个数与第二个参数比较返回小的
print(result)
#     接受的两个参数，也可以大小一致;第二个参数只是一个单独的值时，其实是用到了维度的广播机制；
result = np.maximum( [-2, -1, 4, 1, 2], [1, 2, 3, 4, 5] )
print(result)

# 6. 求平均值
result = np.mean(score) # 获取所有数据的平均值
print(result)
result = np.mean(score,axis=0) # 获取某一行或者某一列的平均值
print(result)

# 7. 求前缀和
arr	= np.array([[1,2,3], [4,5,6]])
print(arr)
print(arr.cumsum(0)) # 按轴0方向累加前面的值
'''
[1, 2, 3]------>|1|2|3|
[4, 5, 6]------>|1+4=5|2+5=7|3+6=9|
'''
print(arr.cumsum(1)) # 按轴1方向累加前面的值
'''
[1, 2, 3]------>|1|2+1=3|3+2+1=6|
[4, 5, 6]------>|4|4+5=9|4+5+6=15|
'''

# 8. argmin求最小值索引
result = np.argmin(score,axis=0)
res=np.min(score,axis=0) # 按轴0找最小值，第一列最小值是75，索引是2，第二列最小值是81，索引是1
print(result,res)

# 9. 求每一列的标准差（这里是总体标准差）
# 标准差一组数据的离散程度。标准差越大，说明这组数越分散，越小说明越聚集
result = np.std(score,axis=0)  # 按轴0计算标准差，第一列的标准差是...第二列的标准差是...
print(result)

# 10. 极差
result = np.ptp(score,axis=None) # 就是最大值和最小值的差
print(result)
```

# 九、数组的拼接
    有的时候我们需要将两个数据加起来一起研究分析，我们就可以将其进行拼接然后分析

## 1、concatenate()   --根据轴连接两个数组，不改变数组的维度

```python
import numpy as np

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print(b,'\n')

# 1. concatenate() 根据轴连接两个数组，不改变数组的维度
print ('沿轴 0 连接两个数组：')
print (np.concatenate((a,b),axis= 0))  # 要求a,b两个数组的维度相同
print ('\n')
print ('沿轴 1 连接两个数组：')
print (np.concatenate((a,b),axis = 1))
print()
```

## 2、stack()  --根据轴进行堆叠，2维会变为3维
    沿哪个轴堆叠，那个轴就会成为新轴，并且其shqpe就等于堆叠的数组个数
    比如 10个3行4列的：
        沿轴0堆叠: 堆叠出来的数组的shape是(10,3,4)
        沿轴1堆叠: 堆叠出来的数组的shape是(3,10,4)
        沿轴2堆叠: 堆叠出来的数组的shape是(3,4,10)

```python
import numpy as np

# 2.  根据轴进行堆叠，2维会变为3维
# 创建一个包含了10个 3行4列的ndarray 的列表
arrays = [np.arange(12).reshape(3, 4) for _ in range(10)]  
print(arrays)
print(len(arrays))

print ('沿轴 0 连接 10个 数组：')
result0 = np.stack(arrays,axis= 0)
print ('拼接后的形状：',result0.shape)  # (10,3,4)
print (result0)
print('-'*50)

print ('沿轴 1 连接 10个 数组：')
print(arrays[0])
result1=np.stack(arrays,axis = 1)
print ('拼接后的形状：',result1.shape)  #(3,10,4)
print (result1)
print('-'*50)

print ('沿轴 2 连接 10个 数组：这里-1和2是等价的')
result2=np.stack(arrays,axis = -1)
print(arrays[0])
print('拼接后的形状：',result2.shape)
print(result2)
```

# 十一、np.inf  和  np.nan
## np.inf
    C 语言中表示最大的正整数值是 0x7FFF FFFF，最小的负整数是 0x8000 0000。
    inf 表示无穷大，需要使用 float(‘inf’) 函数来转化
    对应的有 float('-inf') 表示无穷小了。
    这样你就可以使用任意数来判断和它的关系了。
    那什么时候会出现inf呢？ 
    比如一个数字除以0，Python中会报错，但是numpy中会是一个inf或者-inf
    inf == inf   结果是True
    inf != inf   结果是False
    1/0        结果是 inf
    -1/0       结果是-inf
## np.nan
    nan（not a number）
    这种写法在 pandans 中常见
    表示缺失的数据，所以一般用 nan 来表示
    任何东西与nan做运算  结果都是 nan
    包括 inf*nan 那结果都是nan   无穷都干不过nan
    nan != nan  结果是True   不存在是不等于不存在的，这也很正常，两人的性别未知，你也不能说他们一定是同性
    nan == nan  结果是False
    
## 练习--处理数组中的nan元素
    nan和inf 在numpy中都是float类型
    numpy中 Fasle就是0，True是非零
    所以可以将数组中的nan用 t!=t 替换成False
    然后用np.count_nonzero(t!=t)统计  非0个数  就能得到  非nan的元素个数  了

```python
import numpy as np

# 创建一个nan和inf ，二者其实都是属于float类型
a = np.nan
b = np.inf
print(a,type(a))
print(b,type(b))

# 判断数组中 nan 的元素个数
t = np.arange(24,dtype=float).reshape(4,6)
t[3,4] = np.nan  # 将数组中两个元素改成nan
t[2,4] = np.nan  
print('t数组最初的样子：\n',t)

# 我们可以把以下两个特性结合使用来判断nan的个数
print('t中非零元素个数：\n',np.count_nonzero(t))
print('np.nan==np.na 的判断结果：\n',np.nan==np.nan)  # np.nan != np.nan 结果是TRUE
print('-'*50)

print('将t中所有元素自己和自己比较并生成只有True和False的数组(原数组t没变)：\n',t!=t)  
print('t中非0(非False)元素的个数：\n',np.count_nonzero(t != t)) 


t[np.isnan(t)] = 0
print('将t中 nan 替换为0：\n',t)
print('-'*50)

print('练习：处理数组中nan')
t = np.arange(24).reshape(4,6).astype('float')
t[1,3:] = np.nan  # 将数组中的一部分替换nan
print('原数组t：\n',t)
print('-'*50)
print('t的形状：\n',t.shape)

# 遍历每一列，然后判断每一列是否有nan
for i in range(t.shape[1]):
    temp_col = t[:,i]  # 获取当前列数据
    nan_num = np.count_nonzero(temp_col != temp_col)  # 判断当前列的数据中是否含有nan (和自身不等的就是 nan)
    if nan_num != 0:  # 条件成立说明含有nan
        # 将这一列不为nan的数据拿出来以便计算平均值(和自身相等的就不是 nan)
        temp_col_not_nan = temp_col[temp_col == temp_col]
        print('第{}列中不是nan的元素有（用这些数据计算平均值）：\n'.format(i),temp_col_not_nan)
        temp_col[np.isnan( temp_col )] = np.mean( temp_col_not_nan )  # 将nan替换成这一列的平均值
    print('第{}列处理完毕!'.format(i))

print('将t中所有nan替换成这一列的平均值后的t数组：\n',t)
```

## numpy 里的深拷贝和浅拷贝
    copy() 会返回新数组，两个数组id不一样
    直接赋值的话，两个数组的id一样

```python
import numpy as np

a=np.array([1,2,3])
b=a.copy()  # 深拷贝
c=a        # 浅拷贝

print(a)
print(b)
print(c)
print(id(a))
print(id(b))
print(id(c))
```
